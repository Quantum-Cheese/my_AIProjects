### 调试




#### 1. 尝试调整随机噪点  

Ornstein–Uhlenbeck process，根据维基百科，分布曲线的长期固定方差为：
var = sigma^2/2theta
sigma越大，theta越小，会使得方差越大，产生的随机数变动的幅度也就越大
因此决定往方差更大的方向调整，看会不会促进随机探索

*  sigma = 0.3 theta=0.1  

__Trial_01__ : 明显发现随机探索增强了，刚开始飞出边界后，开始尝试下落和减速，但动作不稳定，到后期就在地面附近上下跳动，不能起飞的情况
记录下了阶段累积奖励 rewards_hover_01.txt 和每个时间步的动作向量 actions_hover_01.txt
也许是因为超出目标高度的惩罚是3次函数，而起飞惩罚只是常数，智能体宁愿停留在地面不动

绘制 action向量（只有force_z）每个时间步的变化曲线

````python
def plot_action_cruve(file_name):
    file=open(file_name,"r")
    actions=[]
    for line in file.readlines():
        action=line.split(",")
        for a in action:
            actions.append(a)
    lst=[]
    for a in actions:
        b=a.strip('[').strip(']').split(" ")
        lst.append(b)
    
    action_values=[]
    for l in lst:
        for s in l:
            if (s!='0.' and s!=""):
                    if s!=']':
                        s=s.rstrip("]")
                        action_values.append(float(s))                  
    import matplotlib.pyplot as plt
    plt.plot(action_values)
````

````python
plot_action_cruve("actions_hover_01.txt")
````

__分析__：action的变化符合Ornstein–Uhlenbeck噪点的设置，开始方差很大，然后逐渐回归平均值（接近0）；表明action主要受到随机OU噪点的支配，没有从环境反馈中学到提升奖励的方法；可能是由于agent学习能力较差，下面着重提升学习性能

#### 7. 更改DDPG算法流程
之前的噪点是在后期把action维度补全时才加上的  

这应该并不合理，会导致真实的动作没有被学习到，模型的动作评估只局限于加噪点之前的动作
* 现在把噪点加在了 act（）方法中，即每次Actor模型根据state选择action后立即加上噪点，使得加入了噪点以后的action会被放入经验缓存区，参与到学习过程
